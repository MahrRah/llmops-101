```python
import os 

from semantic_kernel import Kernel
from semantic_kernel.connectors.ai.open_ai import AzureChatCompletion

kernel = Kernel()

kernel.add_service(AzureChatCompletion(
    deployment_name=os.getenv("DEPLOYMENT_NAME"),
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    base_url=os.getenv("AZURE_OPENAI_ENDPOINT"),
))
```

```python
from semantic_kernel.connectors.ai.chat_completion_client_base import ChatCompletionClientBase

chat_completion_service = kernel.get_service(type=ChatCompletionClientBase)
```

```python
from semantic_kernel.contents.chat_history import ChatHistory
from semantic_kernel.connectors.ai.prompt_execution_settings import PromptExecutionSettings
chat_history = ChatHistory()

chat_history.add_system_message("You are a helpful assistant.")
chat_history.add_user_message("Does Azure OpenAI support customer managed keys?")
chat_history.add_assistant_message("Yes, customer managed keys are supported by Azure OpenAI.")


chat_history.add_user_message("Do other Azure AI services support this too?")

response =  await chat_completion_service.get_chat_message_content(
    settings = PromptExecutionSettings(),
    chat_history=chat_history,
    kernel=kernel,
)

print(response.content)
    
```
